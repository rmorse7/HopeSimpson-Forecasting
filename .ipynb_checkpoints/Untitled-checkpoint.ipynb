```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from scipy import signal
```

```python
data = pd.read_csv('data/TexasData.csv')
```

```python
data.head()
```

```python
def date_address(date):
    dates = date.split("/")
    first = dates[0]
    if len(first) == 4:
        dates[0] = first[2:]
    return "/".join(dates)

data["Date"] = data["Date"].map(date_address)
```

```python
data =data.T
```

```python
data.rename(columns={'Date':'County'})
data.to_csv("TexasDataTransposed.csv", index = True)
```
