# Dataset - US

```python
import pandas as pd
```

```python
srcUS = "../time_series_covid19_confirmed_US.csv"
dest = "TexasData.csv"
tct = 'Texas_Counties.csv'

# Read data
usDf = pd.read_csv(srcUS)
tcounties = pd.read_csv(tct)
tcounties = tcounties['County']
tcounties = [c.split(' County')[0] for c in tcounties]

texasDf = usDf[usDf['Province_State'] == 'Texas']

outdf = {}
dates = texasDf.columns.values.tolist()[13:]
outdf['Date'] = dates

for county in tcounties:
    row = texasDf[texasDf['Admin2'] == county].values.tolist()[0]
    data = row[13:]
    outdf[county] = data

outdf = pd.DataFrame(data = outdf)
print(outdf.head(10))
```

```python
# Save transformed dataset
outdf.to_csv(dest, index = False)
```
